# DataSet
Each record in the database describes a Boston suburb or town. The data was drawn from the Boston Standard Metropolitan Statistical Area (SMSA) in 1970. The attributes are deﬁned as follows (taken from the UCI Machine Learning Repository1): CRIM: per capita crime rate by town

1. ZN: proportion of residential land zoned for lots over 25,000 sq.ft.  
2. INDUS: proportion of non-retail business acres per town  
3. CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)  
4. NOX: nitric oxides concentration (parts per 10 million)  
5. RM: average number of rooms per dwelling  
6. AGE: proportion of owner-occupied units built prior to 1940  
7. DIS: weighted distances to ﬁve Boston employment centers  
8. RAD: index of accessibility to radial highways  
9. TAX: full-value property-tax rate per \$10K   
10. PTRATIO: pupil-teacher ratio by town  
11. B: 1000(Bk−0.63)2 where Bk is the proportion of blacks by town   
12. LSTAT: % lower status of the population  
13. MEDV: Median value of owner-occupied homes in $1000s  


```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
housing_df = pd.read_csv("./housing.csv", header=None, delimiter=r"\s+", names=column_names)
housing_df.head(10)
```

## Important Columns
We see that CRIM, RM, LSTAT are columns which have a high correlation or rather they seem to have a polynomial relation.
So let's try to build a model on these columns first rather than taking the whole dataset.

# Scaling The Columns

```python
from sklearn.preprocessing import StandardScaler
```

```python
inputs = housing_df.drop('MEDV', axis=1)
scaler = StandardScaler().fit(inputs)
```

### Creating Input/Target arrays

```python
scaled_inputs = scaler.transform(inputs)
```

```python
print(pd.DataFrame(scaled_inputs).describe())
scaled_inputs
```

```python
sns.boxplot(data=pd.DataFrame(scaled_inputs))
```

```python
targets= housing_df['MEDV']
```

# Train/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
inputs_train , inputs_test , targets_train , targets_test = train_test_split(scaled_inputs , targets , test_size=0.25,random_state=4)
```

```python
pd.DataFrame(inputs_test)[[0 , 5 , 12]].to_numpy().shape
```

Let's also create an array with CRIM, RM, LSTAT columns as features.

```python
nw_input_train , nw_input_test = pd.DataFrame(inputs_train)[[0 , 5 , 12]].to_numpy() , pd.DataFrame(inputs_test)[[0 , 5 , 12]].to_numpy()
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error
```

#### Model With Only Important Columns

```python
#instantiate the model
model = LinearRegression()

#fit the model on trianing data
model.fit(nw_input_train , targets_train)

#predictions
preds = model.predict(nw_input_test)
# loss
var = explained_variance_score(targets_test , preds)
loss = np.sqrt(mean_squared_error(targets_test , preds))
print('Variance Score : ' , var)
print('RMSE loss: ', loss)
```

#### Linear Model With All Features

```python
#instantiate the model
model = LinearRegression()

#fit the model on trianing data
model.fit(inputs_train , targets_train)

#predictions
preds = model.predict(inputs_test)
# loss
var = explained_variance_score(targets_test , preds)
loss = np.sqrt(mean_squared_error(targets_test , preds))
print('Variance Score : ' , var)
print('RMSE loss: ', loss)
```

The loss in the model with all features is higher, which should be quite obvious, but as a ML engineer it's your job to verify the obvious too.
